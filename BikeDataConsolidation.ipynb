```python
import pandas as pd
```

```python
file = "./2018-citibike-tripdata.csv"
statname = []
statlat = []
statlon = []
for i in range(1,13):
    if i < 10:
        file = f"./20180{i}-citibike-tripdata.csv"
    else:
        file = f"./2018{i}-citibike-tripdata.csv"
    tripdatadf = pd.read_csv(file)
    for index, row in tripdatadf.iterrows():
        try:
            idx = statname.index(row["start station name"])
        except ValueError:
            statname.append(row["start station name"])
            statlat.append(row["start station latitude"])
            statlon.append(row["start station longitude"])
        try:
            idx = statname.index(row["end station name"])
        except ValueError:
            statname.append(row["end station name"])
            statlat.append(row["end station latitude"])
            statlon.append(row["end station longitude"])
for i in range(1,3):
    file = f"./201804-citibike-tripdata_{i}.csv"
    tripdatadf = pd.read_csv(file)
    for index, row in tripdatadf.iterrows():
        try:
            idx = statname.index(row["start station name"])
        except ValueError:
            statname.append(row["start station name"])
            statlat.append(row["start station latitude"])
            statlon.append(row["start station longitude"])
        try:
            idx = statname.index(row["end station name"])
        except ValueError:
            statname.append(row["end station name"])
            statlat.append(row["end station latitude"])
            statlon.append(row["end station longitude"])
```

```python
stationsdict = {"Station Name":statname,
                "Station Latitude":statlat,
                "Station Longitude":statlon}
stationsdf = pd.DataFrame(stationsdict)
stationsdf.to_csv("./stationlocations.csv",header=True,index=False)
```

```python
file = "./2018-citibike-tripdata.csv"
finaldf = pd.DataFrame(columns=["ride_id","rideable_type","started_at","ended_at","start_station_name","member_casual","trip_duration"])
for y in range(2024,2025):
    for m in range(1,13):
        if m < 10:
            month = f"0{m}"
        else:
            month = f"{m}"
        for f in range(1,2):
            file= f"{y}{month}-citibike-tripdata_{f}.csv"
            try:
                testdf = pd.read_csv(file)
                trimtest = testdf[["ride_id","rideable_type","started_at","ended_at","start_station_name","member_casual"]]
                trimtest["started_at"] = pd.to_datetime(trimtest["started_at"])
                trimtest["ended_at"] = pd.to_datetime(trimtest["ended_at"])
                trimtest["trip_duration"]= trimtest["ended_at"]-trimtest["started_at"]
                finaldf = pd.concat([finaldf,trimtest],ignore_index = True)
            except:
                continue
            
finaldf.info()
```

```python
finaldf.to_csv("./biketype.csv",header=True,index=False)
```
